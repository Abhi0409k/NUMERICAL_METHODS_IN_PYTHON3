```python
# Applicable only for positive definite matrix
# in real matrix i.e A = A(Transpose) ie symmetric
# At anypoint if the elements turn out to be complex or invalid this algorithm fails
# check before using

#(0)
# importing required libraries
import numpy as np
import sys

# (1)
# taking inputs
# creating matrices
# validating the inputs

n = int(input('Enter the size of the matrix \t'))
X = np.zeros(n)
Y = np.zeros(n)
L = np.zeros((n,n))

print('Enter the entries of A matrix in a single line separated by a space')
entries = list(map(float, input().split()))
A = np.array(entries).reshape(n,n)
    
# (2)
# Cholesky decomposition

for i in range (n):
    for j in range(i+1):
        Val = 0
        
        # for diagonal elements
        if j == i:
            for k in range(j):
                Val = Val + np.square(L[j,k])
            L[j,j] = np.sqrt(A[j,j]-Val)
        
        # for lower triagular elemets of L ie all elements
        else :
            for k in range(j):
                Val = Val + (L[i,k]*L[j,k])
            if(L[j,j] > 0):
                L[i,j] = (A[i,j]-Val)/L[j,j]

print(L)
U = np.transpose(L)

# (3)
# taking values for B and calculating the solution 
# Back substitution twice once for LY = B then Y = UX

print('Enter the entries of B matrix in a single line separated by a space')
entries = list(map(float, input().split()))
B = np.array(entries).reshape(n,1)
            
L1 = np.concatenate((L,B), axis=1)    # for lower traingular matrix
for i in range(n):
    Y[i] = L1[i,n]
    for j in range(i-1,-1,-1):
        Y[i] =Y[i] - L1[i,j]*Y[j]
    Y[i] = Y[i]/L1[i,i]

Y1 = np.reshape(Y, (n, 1))
U1 = np.concatenate((U,Y1), axis=1)    # for upper traingular matrix
for i in range(n-1,-1,-1):
    X[i] = U1[i,n]
    for j in range(i+1,n):
        X[i] =X[i] - U1[i,j]*X[j]
    X[i] = X[i]/U1[i,i]
    
# (4)
# Displaying Solution

print('The solution vector is:')
for i in range(n):
    print('X[%d] =%0.4f' %(i,X[i]),'\t')
```
